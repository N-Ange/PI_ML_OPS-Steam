```python
ruta_users = 'data\oaustralian_users_items.json'

# Se lee de cada línea del dataset
filas_users = []
with open(ruta_users, encoding='utf-8') as f:
    for line in f.readlines():
        filas_users.append(ast.literal_eval(line))

# Se convierte en dataframe
df_users = pd.DataFrame(filas_users)
df_users
```
