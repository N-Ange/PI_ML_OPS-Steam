```python
#se crea un dataframe usaniendo los dataframe df_games y df_reviews para realizar un sistema de recomendaciones
df_reviews_year = pd.merge(df_games,df_reviews.rename(columns= {"reviews_item_id":"item_id"}),on = "item_id",how = "inner" )
```

```python
df_reviews_year
```

```python
#se elimnan las columnas que no se van a usar
columnas = ["developer", "user_id", "price", "genres", "early_access" ,"publisher", "app_name","developer","user_url", "release_year", "reviews_helpful"]
df_reviews_year.drop(columnas, axis = 1, inplace= True)
```

```python
df_reviews_year
```

```python
# se eliminan los que no tienen fecha de review
df_reviews_year = df_reviews_year[df_reviews_year["reviews_posted"] != "Formato invalido"]

```

```python
df_reviews_year
```

```python
# Se extrae el aÃ±o de la fecha de posteo de la review

df_reviews_year["reviews_posted"] = pd.to_datetime(df_reviews_year["reviews_posted"]).dt.year 
```

```python
df_reviews_year["reviews_posted"].dtype
```

```python
df_reviews_year
```

```python
#se almacena el dataframe en un archivo parquet
ruta = "data/df_reviews_year.parquet"
table = pa.Table.from_pandas(df_reviews_year)
pq.write_table(table,ruta)
```
