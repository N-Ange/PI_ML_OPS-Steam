```python
ruta_review = 'data/australian_user_reviews.json'

# Se lee de cada línea del dataset
filas_review = []
with open(ruta_review, encoding='utf-8') as f:
    for line in f.readlines():
        filas_review.append(ast.literal_eval(line))

# Se convierte en dataframe
df_reviews = pd.DataFrame(filas_review)
df_reviews
```

```python
#Se observa que los comentarios son los mismos, asique se procede a eliminar los duplicados

df_reviews = df_reviews.drop_duplicates(subset='user_id', keep='first')
functions.duplicados(df_reviews, 'user_id')
df_reviews.reset_index(drop=True)
```

```python
# Se revisa como esta compuesto review
df_reviews['reviews'][0]
```

```python
#Se observa que es una columna anidada, se busca generar un dataframe con estos datos.
df_reviews2 = pd.json_normalize(df_reviews["reviews"])
df_reviews2 

```

```python
#Con esta transformacion se pierden el user_id y user_url, pero mantienen la posicion, por lo tanto se puedem concatenar.

df_reviews2 = pd.concat([df_reviews[["user_id", "user_url"]], df_reviews2],axis=1)
df_reviews2
```

```python
#Se utiliza pd.melt para hacer filas con los registros de la columna
df_reviews2 = pd.melt(df_reviews2,id_vars =["user_id","user_url"],
                     value_vars=list(range(9)),
                     value_name="reviews")
```

```python
#Se observan valores None, lo cual se decide eliminar.
df_reviews2.dropna(inplace=True)
```

```python
#se separan por columnas las reviews
df_reviews = df_reviews2["reviews"].apply(pd.Series,dtype="object")
df_reviews = df_reviews.add_prefix('reviews_')
```

```python
#Se agregan user_id y user_url
df_reviews = pd.concat([df_reviews2[["user_id", "user_url"]], df_reviews],axis = 1)
```

```python
#Hay espacios vacios que no son nulos, probablemente sean " "
df_reviews["reviews_last_edited"][0]
```

```python
#se remplazan los espacios " " por None
df_reviews.replace('',None, inplace= True)
```

```python
# se revisan los tipos de datos
functions.tipo_dato(df_reviews)
```

```python
# Hay un gran faltante de datos en reviews_funy y reviews_last_edited (mayor al 85%) por lo tanto se decidio eliminar esas columnas

df_reviews.drop(columns = ["reviews_funny" , "reviews_last_edited"], inplace=True)
```

```python
#Se necesita que la fecha de la review tenga format YYYY-MM-DD. Se van a usar expresiones regulares para extrar la infromacion.
df_reviews["reviews_posted"] = df_reviews["reviews_posted"].apply(functions.conversion_fecha)
```
