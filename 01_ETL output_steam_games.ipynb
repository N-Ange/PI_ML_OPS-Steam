Se importan las librerias que se van a usar en el codigo

```python
import pandas as pd
import json

%load_ext autoreload
%autoreload 2
import functions

import warnings
warnings.filterwarnings("ignore")
```

Se extrae la informacion del DataSet para almacenarla en un Dataframe

```python
#Ruta del dataset.

ruta_DS =  "data\output_steam_games.json"

```

```python
#Lectura de las lineas del dataset.
rows_games = []
with open(ruta_DS) as f:
    for line in f.readlines():
        data = json.loads(line)
        rows_games.append(data)
```

```python
#Convercion en dataframe
df_games = pd.DataFrame(rows_games)
```

```python
df_games
```

Se empieza a trabajar en la limpieza del Dataframe

```python
#Eliminar filas completamente con Nullos
df_games = df_games.dropna(how="all").reset_index(drop=True)
```

```python
#Revistar estdadisticas del dataframe

functions.tipo_dato(df_games)
```

```python
#Buscar duplicados
functions.duplicados(df_games, "id")
```

```python
# Eliminar duplicado
df_games.drop(index = 13894, inplace= True)
```

```python
#revisar segundo duplicado, buscando informacion del developer

df_games[df_games["developer"] == "Rocksteady Studios,Feral Interactive (Mac)"]
```

```python
#Al revisar que habia ya estaba cargado el registro procedo a eliminarlos.
drop_index = [74,30961] 
df_games.drop(drop_index, inplace= True)
```

```python
#Se resetan los indices
df_games.reset_index(drop=True)
```

```python
#Se vuelve a revisar que no queden duplicados
functions.duplicados(df_games, "id")
```

```python
# Se crea la columna con el año de lanzamiento.
df_games["release_year"] = df_games["release_date"].apply(functions.obtener_anio)
```

```python
# se elimina la columna release_date
df_games.drop("release_date",axis = 1, inplace= True)
```

```python
#Las columnas url y reviews_url no se van a estar usando, asique se decidio eliminarlas.

df_games.drop(["url","reviews_url"],axis =1, inplace = True)
```

```python
# En las columnas publisher', 'app_name', 'title', 'developer' se remplazan por valores "No disponible"
columnas = ["publisher", "app_name", "title" , "developer"]
df_games =  pd.concat([df_games.drop(columnas, axis=1), df_games[columnas].fillna('No disponible')], axis=1)
```

```python
#Se nesesita trabjar con la columna price, pero al tener casos con valores gratuitos o Free to play, se decide remplazar esos valores por 0

df_games["price"] = df_games["price"].apply(functions.remplazar_precio) 
```

```python
df_games
```

```python
#Se eliminan las columnas tags y specs por que no se consideran necesarias para el modelo.
df_games = df_games.drop(["tags","specs"], axis =1)
```

```python
functions.tipo_dato(df_games)
```

Luego de finalizar la limpieza de datos, se guarda la informacion en CSV.

```python
#La columna genres esta formada por una lista, Se nesesita hacer una fila por cada genero.
df_games =df_games.explode("genres")
df_games = df_games.dropna(subset=["genres"])
df_games
```

```python
df_games_location = "data/output_steam_games.csv"
df_games.to_csv(df_games_location, index=False, encoding="utf-8")
print(f"El archivo fue guardado correctamente")
```
